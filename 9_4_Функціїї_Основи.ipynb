2. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `calculation` —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ 4 –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É int - `a`, `b`, `c`, `d` —ñ –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–∞–∫–æ—ó —Ñ–æ—Ä–º—É–ª–∏: $ (a + b - c) * d $.

  –í–∏–∫–ª–∏—á—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤ —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω:
  ```
  a = 16
  b = 4
  c = 9
  d = 3
  ```

````python
a = 16
b = 4
c = 9
d = 3

def calculation(a, b, c, d):
    return (a + b - c) * d

calculation(a, b, c, d)
````

4. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers`, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–≤–∞ —á–∏—Å–ª–∞ —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —ñ –ø–æ–≤–µ—Ä—Ç–∞—î —ó—Ö –¥–æ–±—É—Ç–æ–∫.

  –í–∏–∫–ª–∏—á—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers` –Ω–∞ –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª
- 10 —ñ 2
- 7 —ñ 8
- 1024 —ñ 999.
  
  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–æ–∂–Ω–æ–≥–æ –≤–∏–∫–ª–∏–∫—É.
  –ü—Ä–∏ –≤–∏–∫–æ–Ω–∞–Ω–Ω—ñ —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü–∏–∫–ª—É - –æ—Ç—Ä–∏–º–∞—î—Ç–µ 2—Ö –±–∞–ª—ñ–≤ –∑–∞ –Ω—å–æ–≥–æ :)

````python
def multiply_numbers(a, b):
     return a * b

pairs = [(10, 2), (7, 8), (1024, 999)]

for a, b in pairs:
    multiply = multiply_numbers(a, b)
    print(f'{multiply}')
````

    –Ø–∫—â–æ —Ç–∏–ø–∏, –Ω–∞–¥–∞–Ω—ñ —É —Ñ—É–Ω–∫—Ü—ñ—é, –Ω–µ `int` - —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏–≤–æ–¥–∏—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.".

  2. –ó–∞–ø—É—Å—Ç—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª:
- 12 —ñ 6

- 11 i 1.1

    —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: " - —ñ –¥–∞–ª—ñ —Ç–µ, —â–æ –ø–æ–≤–µ—Ä–Ω—É–ª–∞ —Ñ—É–Ω–∫—Ü—ñ—è.

  3. –î–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: "–ß–æ–º—É –≤ –¥–µ—è–∫–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä–Ω—É–ª–∞ None?"

````python
pairs = [(12, 6), (20, 0), (112, 10), (99, 0), (11, 1.1)]
````

````python
def int_division(a: int, b: int) -> int:
  assert type(a) == int and type(b) == int
  if b != 0:
    return a/b
  else:
    return "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0"

for a, b in pairs:
    result = int_division(a, b)
    print(result)
````

````python
def int_division(a, b):
  if b != 0:
    return a/b
  else:
    return "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0"

for a, b in pairs:
  if type(a) == int and type(b) == int:
    result = int_division(a, b)
    print(result)
  else:
    print('–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.')
````
